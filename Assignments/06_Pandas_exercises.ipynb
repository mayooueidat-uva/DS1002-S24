11.  What is the correlation between the variables `sepal_length` and `sepal_width` for all the irises?

```python
# finding correlations between sepal length and width for setosa irises...
# ...using the .corr() function
iris_setosa[["sepal_length", "sepal_width"]].corr()
# correlation between sepal_length and sepal_width is 0.74.
```

```python
# dataframe with ONLY virginica in it
iris_virginica = iris_w_idx.loc['virginica']
# showing the first five rows of the dataframe to make sure...
iris_virginica[:5]
```
